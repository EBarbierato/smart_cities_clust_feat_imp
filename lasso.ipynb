# Lasso

```python
import pandas as pd
from sklearn.linear_model import Lasso

# Load the dataset
file_path = 'Smart_City_index_headers.csv'
data = pd.read_csv(file_path)

# Prepare the data
X = data.drop(columns=['Id', 'City', 'Country', 'SmartCity_Index', 'SmartCity_Index_relative_Edmonton'])
y = data['SmartCity_Index']

# Fit Lasso model
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)

# Display and sort feature weights
feature_weights = pd.Series(lasso.coef_, index=X.columns)
feature_weights_sorted = feature_weights.sort_values(ascending=False)
print("Lasso feature weights (sorted):\n", feature_weights_sorted)
```

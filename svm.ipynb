```python
import pandas as pd
from sklearn.svm import SVR

# Load the dataset
file_path = 'Smart_City_index_headers.csv'
data = pd.read_csv(file_path)

# Prepare the data
X = data.drop(columns=['Id', 'City', 'Country', 'SmartCity_Index', 'SmartCity_Index_relative_Edmonton'])
y = data['SmartCity_Index']

# Fit Linear SVM model
svr = SVR(kernel='linear')
svr.fit(X, y)

# Display and sort feature weights
feature_weights = pd.Series(svr.coef_[0], index=X.columns)
feature_weights_sorted = feature_weights.sort_values(ascending=False)
print("Linear SVM feature weights (sorted):\n", feature_weights_sorted)
```
